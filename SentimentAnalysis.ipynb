```python
# import libraries
# Load the amazon review dataset
df = pd.read_csv(r'C:\Users\jativerma\reviews.csv')
df.head()
```

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string
# df[['Id', 'ProductId', 'UserId', 'ProfileName', 'HelpfulnessNumerator',
#        'HelpfulnessDenominator', 'Score', 'Time', 'Summary', 'Text']].fillna(0, inplace=True)  
nltk.download('stopwords')
nltk.download('punkt')
 
stop_words = set(stopwords.words('english'))
 
def clean_text(text):
    if isinstance(text, str):  # Check if it's a string
        text = text.lower()
        text = text.translate(str.maketrans('', '', string.punctuation))
        tokens = word_tokenize(text)
        filtered_tokens = [word for word in tokens if word not in stop_words]
        return ' '.join(filtered_tokens)
    else:
        return ''  # Return an empty string for non-string values
 
df['clean_text'] = df['Text'].apply(clean_text)
```

```python
df.dropna(subset=['Score'],inplace=True)
```

```python
df['Score']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
 
# TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=50)  # You can adjust max_features
tfidf_matrix = tfidf_vectorizer.fit_transform(df['clean_text'])
 
# Convert to DataFrame (optional)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
 
# Assume 'labels' contain the sentiment classes (positive, negative, neutral)
X = tfidf_matrix
y = df['Score']
 
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
 
# Initialize and train the model
lr_classifier = LogisticRegression(max_iter=1000)
lr_classifier.fit(X_train, y_train)
 
# Predictions
y_pred = lr_classifier.predict(X_test)
 
# Model evaluation
print(classification_report(y_test, y_pred))
```
